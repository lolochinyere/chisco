```python
# Step 2: Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Upload the dataset
from google.colab import files
uploaded = files.upload()
```

```python
# Replace 'HealthCare Data.csv' with your actual file name
data = pd.read_csv('HealthCare Data.csv')
```

```python
# Step 4: Preliminary Data Inspection
print("Data Overview:")
print(data.head())
print("\nData Info:")
print(data.info())
print("\nMissing Values:")
print(data.isnull().sum())
print("\nDuplicate Entries:")
print(data.duplicated().sum())
```

```python

# Removing duplicates if any
data = data.drop_duplicates()
```

```python
# Treat missing values (for now, we'll drop rows with missing values)
data_cleaned = data.dropna()
```

```python
# Confirming no missing values and no duplicates remain
print("\nData after cleaning:")
print(data_cleaned.info())
print("Duplicate Entries after cleaning:", data_cleaned.duplicated().sum())
```

```python
# Step 5: Statistical Summary
print("Statistical Summary:")
print(data_cleaned.describe())
```

```python
## Step 6: Statistical Summary and Exploratory Data Analysis. Identifying Categorical Variables and Count
categorical_vars = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'thal', 'target']
for var in categorical_vars:
    sns.countplot(x=var, data=data_cleaned)
    plt.title(f'Distribution of {var}')
    plt.show()
```

```python
# Step 7: Study the occurrence of CVD across Age
plt.figure(figsize=(10,6))
sns.histplot(data_cleaned, x='age', hue='target', multiple='stack')
plt.title('CVD Occurrence Across Age')
plt.show()
```

```python
# Step 8: Composition of Overall Patients w.r.t. Gender
plt.figure(figsize=(6,4))
sns.countplot(x='sex', hue='target', data=data_cleaned)
plt.title('CVD Occurrence by Gender')
plt.show()
```

```python
# Step 9: Resting Blood Pressure Analysis
plt.figure(figsize=(8,6))
sns.boxplot(x='target', y='trestbps', data=data_cleaned)
plt.title('Resting Blood Pressure vs CVD')
plt.show()
```

```python
# Step 10: Cholesterol Levels and CVD
plt.figure(figsize=(8,6))
sns.boxplot(x='target', y='chol', data=data_cleaned)
plt.title('Cholesterol Levels vs CVD')
plt.show()
```

```python
# Step 11: Peak Exercise and Occurrence of Heart Attack
plt.figure(figsize=(8,6))
sns.boxplot(x='target', y='thalach', data=data_cleaned)
plt.title('Peak Exercise (Thalach) vs CVD')
plt.show()
```

```python

# Step 12: Thalassemia and CVD
plt.figure(figsize=(8,6))
sns.countplot(x='thal', hue='target', data=data_cleaned)
plt.title('Thalassemia vs CVD')
plt.show()
```

```python
# Step 13: Pair Plot to understand relationships
sns.pairplot(data_cleaned, hue='target', diag_kind='kde')
plt.show()
```

```python
##Step 14: Logistic Regression Model.Build a baseline model to predict using a Logistic Regression and explore the results.
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Features and Target variable
X = data_cleaned.drop(columns=['target'])
y = data_cleaned['target']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Logistic Regression Model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
